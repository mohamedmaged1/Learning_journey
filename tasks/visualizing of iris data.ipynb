```python
df1 = iris.loc[iris["Species"] == 'Iris-setosa']
df2=iris.loc[iris["Species"] == 'Iris-versicolor']
df3=iris.loc[iris["Species"] == 'Iris-virginica']
```

```python
i=0
fig,ax = plt.subplots(2,2,figsize=(15,7))
for feature in ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']:
    i += 1
    plt.subplot(2,2,i)
    sns.distplot(df1[feature], hist=False,label='Iris-setosa')
    sns.distplot(df2[feature], hist=False,label='Iris-versicolor')
    sns.distplot(df3[feature], hist=False,label='Iris-virginica')
plt.show()
```

In the last two pictures, we see that we are easy to distinguish between the blue color and the rest of the colors, meaning it has a distinct distribution from the rest, but the other two types overlap

Also, in the first two pictures, you see that there is an overlap in the three colors, which indicates that there is a common characteristic that they all have together
