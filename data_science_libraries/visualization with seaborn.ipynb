#relplot() :--> enables you to visualize the relationship between twi quantitive variables using either scatter plots or line plots 

we can use relplot instead of scatterplot 

```python
hue_coulor={"Male":"black","Female":"red"}
sns.scatterplot(x=tips.total_bill,y=tips.tip,hue=tips.sex,palette=hue_coulor)
```

```python
hue_coulor={"Male":"black","Female":"red"}
sns.relplot(x="total_bill",y="tip",data=tips,hue="sex",palette=hue_coulor,kind="scatter")
```

```python
hue_coulor={"Male":"black","Female":"red"}
sns.relplot(x="total_bill",y="tip",data=tips,hue="sex",palette=hue_coulor,kind="line")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,row="sex",kind="scatter")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,col="sex",kind="scatter")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,col="sex",row="smoker",kind="scatter")
```

```python
##col_wrap parameter u use it to specify how many col in row 
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",size='size')
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",size='size',hue='size')
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",style='smoker',hue='smoker')
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",alpha=.4)
```

## catplot() --> create different types of categorical plots

```python
sns.countplot(x="smoker",data=tips)
```

```python
sns.catplot(x="smoker",data=tips,kind='count')
```

```python
order=['No','Yes']
sns.catplot(x="smoker",data=tips,kind='count',order=order)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar',ci=None)
```
