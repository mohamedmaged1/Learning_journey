# Bar

```python
x= np.linspace(0,20,20)
y=np.random.randint(20,size=20)
print ('x=',x)
print ('y=',y)
```

```python
plt.bar(x,y,facecolor='r',edgecolor='g',alpha=.5,width=0.9)
```

```python
plt.style.use('bmh')
plt.bar(x,y,color='c',alpha=.5,width=0.9)
plt.text(7, -5, 'Polulation Denisty', style='oblique',
        bbox={'facecolor':'m', 'alpha':0.5, 'pad':10})
for xx ,yy in zip(x,y):
    plt.text(xx,yy,yy,ha='center',va='baseline')
    #for va ['top', 'bottom', 'center', 'baseline', 'center_baseline']
#     for ha ['center',right,left]
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, color='r',alpha = 0.5)
plt.bar(X, -Y2, facecolor='b',alpha = 0.5)
for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')
    plt.ylim(-1.25, +1.25)
```
